```python
import scrapy

class Spider12(scrapy.Spider):
    name = 'spider12'
    allowed_domains = ['pagina12.com.ar']  # Dominios que se van a scrapear
    custom_settings = {
        'FEED_FORMAT': 'json',  # Formato de salida
        'FEED_URI': 'resultados.json',  # Donde se va a gurardar la información
        'DEPTH_LIMIT': 2  # Número de veces que va a correr el proceso
    }

    start_urls = [
        'https://www.pagina12.com.ar/secciones/el-pais',
        'https://www.pagina12.com.ar/secciones/economia',
        'https://www.pagina12.com.ar/secciones/sociedad',
        'https://www.pagina12.com.ar/suplementos/cultura-y-espectaculos',
        'https://www.pagina12.com.ar/secciones/deportes',
        'https://www.pagina12.com.ar/secciones/el-mundo',
        'https://www.pagina12.com.ar/secciones/ciencia'
    ]

    def parse(self, response):
        # Noticia principal
        noticia_principal = response.css('.article-item__content a::attr(href)').get()
        if noticia_principal is not None:
            yield response.follow(noticia_principal, callback=self.parse_nota)

        # Notas semiprincipales
        semi_noticias = response.css('h3.title.ff-26px-w800-ls-07 a::attr(href)').getall()
        for semi_noticia in semi_noticias:
            yield response.follow(semi_noticia, callback=self.parse_nota)

        # Artículos (El resto de noticias)
        articulo_noticias = response.css('h4.title.is-display-inline.ff-22px-w700-ls-07 a::attr(href)').getall()
        for articulo_noticia in articulo_noticias:
            yield response.follow(articulo_noticia, callback=self.parse_nota)

        # Link a la siguiente pagina
        next_page = response.css('div.pagination a.next::attr(href)').get()
        if next_page is not None:
            yield response.follow(next_page, callback=self.parse)

    def parse_nota(self, response):
        # Parseo de la nota
        titulo = response.css('h1::text').get()
        cuerpo = ''.join(response.css('.article-main-content.article-text *::text').getall())
        yield {
            'url': response.url,
            'titulo': titulo,
            'cuerpo': cuerpo
        }
```
