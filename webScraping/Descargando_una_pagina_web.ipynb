```python
from bs4 import BeautifulSoup
```

```python
s = BeautifulSoup(pl12.text,'lxml')
```

```python
# En la variable s se se guarda la sopa que es el texto que esta en pl12, lxml es un parser que guarda el texto largo en pedasitos
```

```python
type(s)
```

```python
print(s.prettify())
```

```python
# Imprime el texto pritificado, nos muestra una intuicion de como está estructurada la página.
```

```python
s.find('ul')
```

```python
# El método find devuelve el primer elemento que encuentre segun le le indiquemos en este caso una etiqueta ul
# con f12 se despliega el inspector de elementos en el navegador
```

```python
s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'})
```

```python
# la funcion s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}) nos devuelve el ul especifico, 
# debemos agregarle el metodo attrs atributo y debemos pasarl un diccionario con el nombre del atributo y valor que esperamos optener.
```

```python
s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
# Agregando find_all('li') optenemos todos los li de la sección.
```

```python
secciones = s.find('ul', attrs={'class':'horizontal-list main-sections hide-on-dropdown'}).find_all('li')
```

```python
seccion = secciones[0]
seccion
```

```python
# seeciones[0] muestra el primer elemento
```

```python
seccion.find('a') 
# muestra la primera etiqueta a
```

```python
seccion.a
# es la versión simplificada de el comando anterior
```

```python
seccion.a.get('href')
# Obtener en forma de string el texto del atributo href
```

```python
seccion.a.get_text()
# Para obtener solo el texto de la etiqueta
```

```python
links_secciones = [seccion.a.get('href') for seccion in secciones]
# Una lista por comprensión de seccion en secciones para obtener todos los links de la clase
```

```python
links_secciones
```

```python
sec = links_secciones[0]
sec
# El primer link de las secciones
```

```python
sec = requests.get(links_secciones[0])
# Enviar un peticion al primer link de la lista 
```

```python
sec
```

```python
sec.status_code
```

```python
s_seccion = BeautifulSoup(sec.text, 'lxml')
# Una soup o sopa de  el primer link o de sec
```

```python
print(s_seccion.prettify())
# Imprimir de forma pritificada
```

```python
top_article = s_seccion.find('div', attrs={'class':'article-item__content'})
top_article                                          
```

```python
top_article.a.get('href')
```

```python
link_top_article = top_article.a.get('href')
link_top_article
```

```python
list_h3 = s_seccion.find_all('h3', attrs={'class':'title ff-26px-w800-ls-07'})
list_links_seccion1 = [link.a.get('href') for link in list_h3]
                   
```

```python
list_links_seccion1
```

```python
list_h4 = s_seccion.find_all('h4', attrs={'class':'title is-display-inline ff-22px-w700-ls-07'})
list_links_seccion2 = [link.a.get('href') for link in list_h4]
```

```python
list_links_seccion2
```

```python
all_links = [link_top_article] + list_links_seccion1 + list_links_seccion2
all_links
```
