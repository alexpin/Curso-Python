```python
import numpy as np
import matplotlib.pyplot as plt

# Definicion de la resolucion de las funciones
N=100

# Definicion de las varibles.
x = np.linspace(-10, 10, num=N)
#y = g(x)

# Definir la primera funcion
def g(x):
  return x**2
# Regresa una parabola

# Definir la segunda funcion
def f(x):
  return np.sin(x)
# Grafica una funcion seno

# Evalua la salida de la funcion, como argumento de otra funcion
f_o_g = f(g(x))
# lo que recibe como argumento es la salida de g(x), esto es y


plt.plot(x, f_o_g);
plt.grid()

# La funcion de la derecha es quien recibe la x.
# La funcion de la Ixquierda sera quien reciba la salida de nuestra segunda funcion
```
