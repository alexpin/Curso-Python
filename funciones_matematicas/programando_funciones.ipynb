```python
import matplotlib.pyplot as plt
import numpy as np
```

Funciones algebraicas

Funcion lineal
Tiene la forma de f(x) = mx + b myb pertenecen a los reales
m=y2−y1/x2−x1
b es el punto de corte en el eje y

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)
y = f(x)

x.shape
y.shape

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
fig.show()

```

Funciones Polinomicas


```python
def f(x):
    return (2*x**5)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```

Funciones potencia, son un caso de las funciones polinomicas

```python
def f(x):
    return x**7

y = f(x)

plt.plot(x,y)
plt.grid()
```

Funciones trigonometricas

```python
def f(x):
  return np.cos(x)

y = f(x)

plt.plot(x,y);
plt.grid()
```

```python
def f(x):
  return np.sin(x)

y = f(x)

plt.plot(x,y);
plt.grid()
```

```python
def f(x):
  return np.tan(x)

y = f(x)

plt.plot(x,y);
plt.grid()
```

Funcion Exponencial

```python
def f(x):
  return np.exp(x)

y = f(x)

plt.plot(x,y);
plt.grid()
```

**Funcion logaritmo**

```python
def f(x):
  return np.log2(x)

x = np.linspace(0.001, 256, num=1000)
y = f(x)

plt.plot(x,y);
plt.grid()
```

```python
def H(x):
  Y = np.zeros(len(x))
  for idx,x in enumerate(x):
    if x >= 0:
      Y[idx]=1
  return Y
 

N = 1000

x = np.linspace(-10,10, num=N)

y = H(x)

plt.plot(x,y);
plt.grid()

```
