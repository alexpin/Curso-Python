Funcion lineal
Tiene la forma de f(x) = mx + b myb pertenecen a los reales
m=y2−y1/x2−x1
b es el punto de corte en el eje y

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)
y = f(x)

x.shape
y.shape

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
fig.show()

```
