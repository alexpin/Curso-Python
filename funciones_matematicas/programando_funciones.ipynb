```python
import matplotlib.pyplot as plt
import numpy as np
```

Funciones algebraicas

Funcion lineal
Tiene la forma de f(x) = mx + b myb pertenecen a los reales
m=y2−y1/x2−x1
b es el punto de corte en el eje y

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)
y = f(x)

x.shape
y.shape

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()
fig.show()

```

Funciones Polinomicas


```python
def f(x):
    return (2*x**5)-(x**4)+(3*x**2)+4

y = f(x)

plt.plot(x,y)
plt.grid()
```

Funciones potencia, son un caso de las funciones polinomicas

```python
def f(x):
    return x**7

y = f(x)

plt.plot(x,y)
plt.grid()
```
