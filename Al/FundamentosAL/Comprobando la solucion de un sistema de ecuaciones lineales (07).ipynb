```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5,5) # Ecalue la funcion entre -5 y 5
```

```python
y_1 = 3*x + 5  # Calcular efectivamente los puntos
y_2 = 2*x + 3
```

```python
plt.figure() # Generar una figura

plt.plot(x, y_1) # los valores que debe contener la figura
plt.plot(x, y_2)

plt.xlim(-5,5) # Definir entre que valores nos muestre cada eje
plt.ylim(-5,5)

plt.axvline(x=0, color='grey') # Coordenadas en las que queramos que se vean la lineas
plt.axhline(y=0, color='grey')
```
