### Importamos las librerias necesarias para el analisis


```python
import pandas as pd
import numpy as np
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded["cancerdata.csv"]))
df
```

```python
X = df.iloc[:,2:]
Y = df.iloc[:,1]
```

```python
X.head()
```

```python
Y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.3, random_state= 4)
modelo = DecisionTreeClassifier()
modelo.fit(X_train, Y_train)
resultado = modelo.score(X_test, Y_test)
print(resultado)
```

```python
# Validacion curxada k fold
from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier()
kfold_validacion = KFold(10)
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, X, Y, cv = kfold_validacion)
print(resultados)
resultados.mean()
```
