```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

bins = Cuantas barras va a agrupar esta frecuencia
cumulative = Acumula los valores en cada barra

```python
sns.histplot(data = tips, x='tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

Diagrama de densidad de como se estan comportando las propinas tanto para males and famele
shade -> completa la figura rellenandola con colores.

```python
sns.kdeplot(data=tips, x ='tip', hue='sex', cumulative=False, shade=True, bw_adjust=1)
plt.show()
```

Grafico escalonado de la proporcion de las propinas a lo largo de su frecuencia

```python
sns.ecdfplot(data=tips, x='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex', kind='hist', multiple='stack')
plt.show()
```
