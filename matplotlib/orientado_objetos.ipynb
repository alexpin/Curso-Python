```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2 
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'b')
fig.show()
```

fig = plt.figure() -> Paso uno definir una figura
las graficas que vamos a pintar se llaman axes
axes = fig.add_axes([0.1,0.1,0.5,0.9]) ->Los dos primeros hacen referencia a la posicion en el lienxo, los dos siguientes hacen referencia al tamano de la figura.

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y, 'b')
axes2.plot(y,x, 'r')
axes2.set_facecolor('gray')
fig.show()
```
