```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

## Histograma
Grafica la ditribucion ordenada de los datos, para la grafica de abajo la mayoria de los datos se concentran en la xona 15 a 16 aproximadamente.

```python
plt.hist(data, bins=10)
plt.show()
```

bins = Numero de barras a la cual va a llegar la distribucion de datos

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

## Boxplot
Tambien conocido como grafico de cajas y bigotes
vert=False -> Cambia la ubicacion de la grafica
patch_artist=True -> Para rellenar y dar un percepcion del rango inercuartilico mas sensilla
showfliers=False -> se utilixa para no mostrar los liers, que introdujimos en la primera linea, en True los muestra

```python
data = np.append(data,200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

# Scatter
Grafico de dispersion

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2
colors = np.random.rand(N)
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
