```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country = ['INDIA', 'JAPAN', 'MEXICO', 'COLOMBIA', 'GERMANY']
population = [1000,800,900,1000,300]
ave = np.average(population)
med = np.median(population)
```

```python
plt.bar(country,population, width=0.5, color=['#8ecae6','#219ebc','#023047','#ffb703','#fb8500'])
plt.xticks(np.arange(5),('India','Japon','Mexico','Colombia','Alemania'), rotation=45)
plt.title('Clientes por paises a 2022')
plt.xlabel('Paiz')
plt.ylabel('Numero de clientes')
plt.axhline(y=ave, color='r', linestyle = '--', label = 'Promedio')
plt.axhline(y=med, color='g', linestyle = ':', label = 'Media')
plt.legend(frameon=False)
plt.show()
```
