```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')

```

```python
tips
```

```python
sns.displot(data=tips, x='total_bill')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', y ='tip')
plt.show()
```

Realixa una agrupacion adicional, en este caso si es hombre o mujer.

```python
sns.displot(data=tips, x='total_bill', y ='tip', hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill', hue='sex')
plt.show()
```

kind -> Tipo de grafico con el que quiero trabajar


```python
sns.displot(data=tips, x='total_bill', hue='sex', kind ='kde', legend=True, palette='dark', alpha=0.25)
plt.show()
```
