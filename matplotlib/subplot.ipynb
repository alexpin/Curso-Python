```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

1 = Una fia
2 = Dos columnas
1= Un indice
Esto es para la primera grafica.
La segunda grafica es con el index 2 y es un histograma de y.

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```
